Create a new column 'income_binary' where <=50K is 0 and >50K is 1

```python
income_binary=[i for i in range(48842)]
for x in range(0,48842):
  if (data["income"].iloc[x]=="<=50K"):
    income_binary[x]=0
  else:
    income_binary[x]=1
```

```python
data["income_binary"]=income_binary
data
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+)

```python
age_group=[i for i in range(48842)]
for x in range(0,48842):
  if (data["age"].iloc[x]<30):
    age_group[x]="youth"
  elif (30<(data["age"].iloc[x])<60):
    age_group[x]="adult"
  else:
    age_group[x]="senior"
```

```python
data["age_group"]=age_group
data
```
